```python
import pandas as pd
df=pd.read_csv("data/movies.csv")
```

```python
cols_to_drop = ["status", "tagline", "poster_path", "backdrop_path", "recommendations", 'production_companies']
df.drop(columns=cols_to_drop, inplace=True, errors="ignore")
```

```python
df.info()
```

```python
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")
df = df.loc[df["release_date"].between("2000-01-01", "2025-12-31")]
```

```python
df.columns
```

```python
df['genres'] = df['genres'].str.replace('-', ', ')
```

```python
df.count()
```

```python
df = df.reset_index(drop=True)
```

```python
df.tail()
```

```python
# Keep movies with at least 10 votes
df = df[df['vote_count'] >= 10]
```

```python
df.info()
```

```python
duplicates = df.duplicated(subset=['title', 'release_date'], keep='first')
num_duplicates = duplicates.sum()

if num_duplicates > 0:
    print(f"⚠️ Found {num_duplicates} duplicate rows. Removing them...")
    df = df.drop_duplicates(subset=['title', 'release_date'], keep='first')
else:
    print("✅ No duplicates found.")
```

```python
# Drop rows where critical info is missing
df = df.dropna(subset=['credits', 'genres', 'release_date'])
```

```python
df.tail()
```

```python
df = df.reset_index(drop=True)
```

```python
df.info()
```

```python

df.columns
```

```python
d = df[['id','popularity','keywords','runtime']]
df.drop(columns = d,inplace =True)
```

```python
df = df.rename(columns={'credits': 'cast'})
```

```python
df['cast'] = df['cast'].str.replace('-', ', ')
# We CREATE 'main_actors' here because the App needs it
def get_top_5_actors(cast_string):
    if isinstance(cast_string, str):
        return ', '.join(cast_string.split(', ')[:5])
    return ""
df['main_actors'] = df['cast'].apply(get_top_5_actors)
```

```python
df.info()
```

```python
# We map codes to full English names.
language_map = {
        'en': 'English', 'fr': 'French', 'es': 'Spanish', 'ja': 'Japanese',
        'it': 'Italian', 'de': 'German', 'ko': 'Korean', 'hi': 'Hindi',
        'pt': 'Portuguese', 'zh': 'Chinese (Mandarin)', 'ru': 'Russian',
        'tr': 'Turkish', 'ta': 'Tamil', 'ml': 'Malayalam', 'nl': 'Dutch',
        'sv': 'Swedish', 'pl': 'Polish', 'cn': 'Chinese (Cantonese)', 
        'da': 'Danish', 'ar': 'Arabic', 'te': 'Telugu', 'no': 'Norwegian',
        'th': 'Thai', 'fi': 'Finnish', 'el': 'Greek', 'fa': 'Persian',
        'he': 'Hebrew', 'tl': 'Tagalog', 'cs': 'Czech', 'ro': 'Romanian',
        'hu': 'Hungarian', 'id': 'Indonesian', 'sr': 'Serbian', 'uk': 'Ukrainian',
        'is': 'Icelandic', 'ca': 'Catalan', 'bn': 'Bengali', 'kn': 'Kannada',
        'et': 'Estonian', 'hr': 'Croatian', 'bg': 'Bulgarian', 'ka': 'Georgian',
        'vi': 'Vietnamese', 'eu': 'Basque', 'ms': 'Malay', 'sk': 'Slovak',
        'lt': 'Lithuanian', 'gl': 'Galician', 'ur': 'Urdu', 'mr': 'Marathi',
        'lv': 'Latvian', 'pa': 'Punjabi', 'af': 'Afrikaans', 'bs': 'Bosnian',
        'sl': 'Slovenian', 'mk': 'Macedonian', 'ku': 'Kurdish', 'bo': 'Tibetan',
        'mn': 'Mongolian', 'xx': 'No Language'
    }
    
    # Map the column. If a code is missing (very rare), it becomes 'Other'
df['language'] = df['original_language'].map(language_map).fillna('Other')
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
    # Extracts '2014.0' and converts to '2014' for clean charts
df['year'] = df['release_date'].dt.year.astype(int)
```

```python
df.info()
```

```python
col_drop =df[['original_language','release_date','budget', 'revenue']]
df.drop(columns=col_drop, inplace =True)
```

```python
df.info()
```

```python
output_path = "data/movies_cleaned.csv"
df.to_csv(output_path, index=False)
```
